```python

def convert_3(obj):
    name_list = []
    counter = 0
    #to convert string of list to list
    for i in ast.literal_eval(obj):
        if counter != 3 :
            name_list.append(i['name'])
            counter+=1
        else:
            break
    return name_list
#as the data is soo big we take only first 3 cast memebers
```

```python
def get_Director(obj):
    name_list = []
    #to convert string of list to list
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            name_list.append(i['name'])
            break
    return name_list
        
#we only extract feature that have job = director
```

```python
import numpy as np

def cosine_similarity(vectors):
    """
    Calculate the cosine similarity between all pairs of vectors.
    
    Args:
    vectors (list of np.ndarray): List of input vectors.
    
    Returns:
    np.ndarray: 2D array containing cosine similarities between all pairs of vectors.
    """
    num_vectors = len(vectors)
    similarities = np.zeros((num_vectors, num_vectors))
    for i in range(num_vectors):
        for j in range(num_vectors):
            if i == j:
                similarities[i, j] = 1.0
            else:
                similarities[i, j] = np.dot(vectors[i], vectors[j]) / (np.linalg.norm(vectors[i]) * np.linalg.norm(vectors[j]))
    return similarities
```

```python
similarity = cosine_similarity(vectors)
```
